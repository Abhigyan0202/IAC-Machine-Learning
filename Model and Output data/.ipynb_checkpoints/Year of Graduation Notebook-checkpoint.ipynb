```python
import pandas as pd
```

```python
df = pd.read_excel('05 Input Data/Year of Graduation Data/Final Lead Data.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
#We consider only those entries in which Academic Year or Current Academic Year is given
df_next = df[df['Academic Year'].notna() | df['What is your current academic year?'].notna()]
```

```python
#Renaming 'What is your current academic year?' column
df_next.rename(columns={'What is your current academic year?': 'Current Academic Year'},inplace=True)
```

```python
df_next.head()
```

```python
#We fill the null entries in Current Academic Year column from the Academic Year column
df_next['Current Academic Year'] = df_next['Current Academic Year'].fillna(df_next['Academic Year'])
```

```python
df_next.info()
```

```python
#We drop the Academic Year column as it is not required now
df_next.drop(['Academic Year'],axis=1,inplace=True)
```

```python
#we fill the empty entries in New College Name column with the entries in Colleges
df_next['New College Name'] = df_next['New College Name'].fillna(df_next['Colleges'])
```

```python
df_next.info()
```

```python
#We drop those entries in which we do not have a college name
df_next = df_next.dropna(subset=['New College Name'])
```

```python
df_next.info()
```

```python
#We drop the Colleges column as it not required 
df_next.drop(['Colleges'],axis=1,inplace=True)
```

```python
#We also rename the New College Name column
df_next.rename(columns={'New College Name': 'College'},inplace=True)
```

```python
#Since the course is mentioned to be of 4 years length, we only need to check current academic year and calculate year of graduation
```

```python
df_next['Current Academic Year'].unique()
```

```python
#We need to convert the string values like '3rd Year', 'Final Year', '2nd Year', '1st Year' into numerical values, for that we use a dictionary
dictionary1 = {
    1.0: 1,
    2.0: 2,
    3.0: 3,
    4.0: 4,
    '1st Year': 1,
    '2nd Year': 2,
    '3rd Year': 3,
    'Final Year': 4
}
```

```python
df_next['Current Academic Year Numerical'] = df_next['Current Academic Year'].apply(lambda x: dictionary1.get(x))
```

```python
df_next.head()
```

```python
#For year of graduation, we use the following calculation
#Year of graduation  = Current Calendar Year + (Course Length-Current Academic Year)
import datetime
Current_year = datetime.datetime.now().year
Course_length = 4
```

```python
df_next['Year of graduation'] = df_next['Current Academic Year Numerical'].apply(lambda x: Current_year +(Course_length -x+1))
```

```python
df_next.head()
```

```python
#Saving the dataframe to an excel file
df_next.to_excel('Year of Graduation file.xlsx')
```
