```python
import pandas as pd
```

```python
original_df = pd.read_excel('05 Input Data/Prediction of Placement Status Data/01 Train Data.xlsx')
```

```python
original_df.head()
```

```python
original_df.info()
```

```python
df = original_df
df.info()
```

```python
#Assuming null values as not placed
df['Placement Status'] = df['Placement Status'].fillna('Not placed')
```

```python
#Dropping duplicate values
df_no_duplicates = df.drop_duplicates(['Email ID'],keep='last')
df_no_duplicates.info()
```

```python
#Choosing only the correct columns
df_selected = df_no_duplicates[['College Name','CGPA','Speaking Skills','ML Knowledge', 'Placement Status']]
df_selected.head()
```

```python
df_selected.info()
```

```python
#LabelEncoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df_selected['College Name'])
df_selected.loc[:,'College Names Labelled'] = le.transform(df_selected['College Name'])
df_selected.head()
```

```python
#Choosing the input and target variables
X = df_selected[['CGPA','Speaking Skills', 'ML Knowledge', 'College Names Labelled']]
y = df_selected['Placement Status']
```

```python
#Importing the model
from sklearn import tree
model_tree = tree.DecisionTreeClassifier()
```

```python
#Feeding the values into the model
model_tree.fit(X,y)
```

```python
#Getting the accuracy of the model
model_tree.score(X,y)
```

```python
#Next step is to use the model to update the Test data file
```

```python
test_df = pd.read_excel('05 Input Data/Prediction of Placement Status Data/02 Test Data.xlsx')
test_df.info()
```

```python
#In the next step, we do some data cleaning in order for the Label Encoder to work appropriately. Without doing so,it gives an error
```

```python
mask_m = test_df['College Name'] == 'AP SHAH INSTITUTE OF TECHNOLOGY '
```

```python
test_df['College Name'] = test_df['College Name'].replace(['AP SHAH INSTITUTE OF TECHNOLOGY '],'ap shah institute of technology ')
```

```python
test_df['College Names Labelled'] = le.transform(test_df['College Name'])
```

```python
test_df.head()
```

```python
#Choosing the target variables of test dataframe
X_test = test_df[['CGPA', 'Speaking Skills', 'ML Knowledge', 'College Names Labelled']]
X_test.head()
```

```python
#Getting the predicted values
y_pred = model_tree.predict(X_test)
y_pred
```

```python
#Putting the predicted values into the original test dataframe
test_df['Placement Status'] = y_pred
```

```python
test_df.head()
```

```python
#Saving the model to an excel file
test_df.to_excel('Placement_Status_Model_Output.xlsx')
```
